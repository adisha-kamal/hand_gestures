```python
import mediapipe as mp
import cv2
```

```python
mp_gestures=mp.solutions.hands
mp_draw=mp.solutions.drawing_utils
```

```python
gestures=mp_gestures.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=gestures.process(img1)
    tipid=[4,8,12,16,20]
    lmlist=[]
    if result.multi_hand_landmarks:
        for handlms in result.multi_hand_landmarks:
            for id,lm in enumerate(handlms.landmark):
                cx=lm.x
                cy=lm.y
                lmlist.append([id,cx,cy])
            
            if len(lmlist)==21:
                fingers=[]
                if lmlist[12][1]<lmlist[20][1]:
                        if lmlist[4][1]>lmlist[3][1]:
                            fingers.append(0)
                        else:
                            fingers.append(1)
                else:
                    if lmlist[4][1]<lmlist[3][1]:
                            fingers.append(0)
                    else:
                            fingers.append(1)
                    
                    #all finger except thump
                    for i in range(1,5):
                        if lmlist[tipid[i]][2]>lmlist[tipid[i]-2][2]:
                            fingers.append(0)
                        else:
                            fingers.append(1)

                gesture = "Unknown"

                if fingers == [0,0,0,0,0]:
                    gesture = "FIST"
                elif fingers == [1,0,0,0,0]:
                    gesture = "THUMB UP"
                elif fingers == [1,1,1,1,1]:
                    gesture = "OPEN PALM"
                elif fingers == [0,0,1,1,1]:
                    gesture = "OK SIGN"
                elif fingers == [1,1,1,1,1]:
                    gesture = "OPEN PALM"
                elif fingers == [0,1,1,1,0]:
                    gesture = "TRIANGLE"
                elif fingers == [0,1,1,0,0]:
                    gesture = "PEACE"
                elif fingers == [1,1,1,0,0]:
                    gesture = "HIGH FIVE"
                elif fingers == [0,1,0,0,1]:
                    gesture = "ROCK"
                elif fingers == [1,0,0,0,1]:
                    gesture = "CALL ME"
                elif fingers == [0,1,0,0,0]:
                    gesture = "POINTING"

                cv2.putText(img,gesture,(30, 80),cv2.FONT_HERSHEY_COMPLEX,1.5,(0, 255, 0),3)

            mp_draw.draw_landmarks(img, handlms, mp_gestures.HAND_CONNECTIONS)

    cv2.imshow("Hand Gestures", img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
